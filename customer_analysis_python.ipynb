# Task
Analyze customer data from "/cleaned dataset (1).csv" by cleaning the data, creating an 'Age_Group' column, and visualizing gender distribution and average spending score by age group.

## Load the dataset

### Subtask:
Load the data from the CSV file into a pandas DataFrame.


**Reasoning**:
I need to load the data from the specified CSV file into a pandas DataFrame to start the analysis.



```python
import pandas as pd
df = pd.read_csv('/cleaned dataset (1).csv')
display(df.head())
```

## Clean and prepare the data

### Subtask:
Convert the `Join_Date` column to datetime objects, standardize the values in the `Gender` column, and create a new column `Age_Group` based on the specified age bins.


**Reasoning**:
Convert the 'Join_Date' column to datetime objects, standardize the 'Gender' column, and create the 'Age_Group' column based on the specified age bins and labels.



```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'], format='%m/%d/%Y')
df['Gender'] = df['Gender'].str.capitalize()
age_bins = [0, 18, 25, 35, 50, float('inf')]
age_labels = ['<18', '18-25', '26-35', '36-50', '>50']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
display(df.head())
```

## Display cleaned data

### Subtask:
Display the first 5 rows of the cleaned and prepared DataFrame.


**Reasoning**:
Display the first 5 rows of the DataFrame to verify the data cleaning and preparation steps.



```python
display(df.head())
```

## Visualize data

### Subtask:
Create a pie chart showing the distribution of customers by gender and a bar chart showing the average spending score by age group.


**Reasoning**:
Calculate gender distribution, create a pie chart, calculate average spending score by age group, and create a bar chart to visualize the data.



```python
import matplotlib.pyplot as plt
gender_distribution = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Customers by Gender')
plt.axis('equal')
plt.show()
average_spending_by_age = df.groupby('Age_Group')['Spending_Score'].mean().sort_index()
plt.figure(figsize=(10, 6))
average_spending_by_age.plot(kind='bar', color='skyblue')
plt.title('Average Spending Score by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Spending Score')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
